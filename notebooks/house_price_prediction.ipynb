```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/house_price.csv')
df.head()
```

```python
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in ['street', 'city', 'statezip', 'country']:
    df[col] = le.fit_transform(df[col])
```

```python
df.info()
df.head()
```

```python
df=df.drop(columns='country')
```

```python
df_corr = df.corr()
plt.figure(figsize=(14, 10))
sns.heatmap(df_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.corr()['price'].abs().sort_values(ascending=False)
```

```python
columns_to_drop = ['date', 'yr_renovated', 'street', 'city', 'statezip']
df = df.drop(columns=columns_to_drop)
```

```python
df.head()
```

```python
import scipy.stats as stats
z_scores = stats.zscore(df)
threshold = 3
print('Size before removing outliers:', df.shape)
outliers_df = df[(z_scores > threshold).any(axis=1)]
df = df[(z_scores <= threshold).all(axis=1)]
print('Size after removing outliers:', df.shape)
```

```python
outliers_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
df_scaled = pd.DataFrame(scaler.transform(df), columns=df.columns)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('price', axis=1)
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
models = [
    ('K-Nearest Neighbors', KNeighborsRegressor()),
    ('Linear Regression', LinearRegression()),
    ('Decision Tree', DecisionTreeRegressor()),
    ('Random Forest', RandomForestRegressor())
]
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
for name, model in models:
    print(name)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    print(f"  Mean Squared Error: {mse}")
    print(f"  R-squared: {r2}")
    print(f"  Mean Absolute Error: {mae}")
```
